## 2.3 Titles table

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
```

```python
address = 'C:/Users/Moanesga/Desktop/Data/titles.csv'
titles = pd.read_csv(address)

titles.columns = ['emp_no','title','from_date','to_date']
titles.head(5)
```

```python
#transforming time variables into date format
titles['from_date'] = pd.to_datetime(titles['from_date'])
titles['from_date'].max()
```

```python
titles.describe(include='all')
```

```python
date = pd.Timestamp('2003-08-01')
titles.iloc[:, 3] = pd.to_datetime(titles.iloc[:, 3], errors='coerce').fillna(date)
titles.head(5)
```

```python
titles.describe(include='all')
```

```python
titles_freq = titles.filter(['title','emp_no'], axis=1)
titles_freq_1 =titles_freq.groupby(by=['title']).count()
titles_freq_1 
```

```python
titles_freq_2 = pd.DataFrame(titles_freq_1,columns =['emp_no'] )
titles_freq_2['total'] = len(titles['emp_no'])
titles_freq_2['percentage'] = (titles_freq_2['emp_no']/titles_freq_2['total'])*100
titles_freq_2.head(5)
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import scipy 
from scipy import stats
```

```python
%matplotlib inline
plt.style.use('ggplot')
rcParams['figure.figsize'] = 20, 8

titles_sorted = titles_freq_2.sort_values('emp_no',ascending=False)
ax = titles_sorted['emp_no'].plot(kind='barh',color = 'coral',fontsize=13)
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x)))) 

totals = []
for i in ax.patches:
    totals.append(i.get_width())

total = sum(totals)
for i in ax.patches:

    ax.text(i.get_width()+.3, i.get_y()+.38, \
            str(round((i.get_width()/total)*100, 2))+'%', fontsize=15,
            color='dimgrey')
    
ax.invert_yaxis()

plt.title("How many employees per job title?",fontsize=20, color='black')
ax.set_ylabel("Job Title", fontsize=18)
ax.set_xlabel("Number of employees", fontsize=18)
plt.show()
```

```python
titles['years_in_job'] = (titles['to_date'] - titles['from_date']).astype('timedelta64[Y]')
years_in_job = pd.DataFrame(titles, columns =['title','years_in_job'])
years_in_job.head()
```

```python
years_in_job_1 = years_in_job.groupby(by='title').mean()
years_in_job_1.head(5)
```

```python
%matplotlib inline
plt.style.use('ggplot')
rcParams['figure.figsize'] = 10, 8

ax1 = years_in_job_1.plot(kind='bar',color = 'lightblue',fontsize=13)
plt.title("How many years hold employees each job title?",fontsize=20, color='black')
ax1.set_ylabel("Number of years", fontsize=18)
ax1.set_xlabel("Job title", fontsize=18)


plt.show()
```

```python
times_change_job = pd.DataFrame(titles, columns =['emp_no','title','from_date'])
times_change_job_1= times_change_job.groupby(by=['emp_no']).count() 
times_change_job_1.head()
```

```python
times_change_job_2= times_change_job_1.groupby(by=['title']).count() 
times_change_job_2.head()
```

```python
%matplotlib inline
plt.style.use('ggplot')
rcParams['figure.figsize'] = 8, 8

ax2 = times_change_job_2['from_date'].plot(kind='bar',color ='salmon',fontsize=13)
plt.title("How many job titles have employees had?",fontsize=20, color='black')

ax2.set_ylabel("Number of employees", fontsize=18)
ax2.set_xlabel("Number of times changing job title", fontsize=10)

totals1 = []
for a in ax2.patches:
    totals1.append(a.get_height())
    
total1 = sum(totals1)  

for a in ax2.patches:
    ax2.text(a.get_x()-.03, a.get_height()+.5, \
            str(round((a.get_height()/total1)*100, 2))+'%', fontsize=15,
                color='dimgrey')

plt.show()
```
